```python
def map_feature(X1, X2, degree=6):
    features = []
    for d in range(degree + 1):
        for i in range(d + 1):
            features.append((X1 ** (d - i)) * (X2 ** i))
    return np.array(features).T
```

```python
print("Original shape of data:", X_train.shape)

mapped_X =  map_feature(X_train[:, 0], X_train[:, 1])
print("Shape after feature mapping:", mapped_X.shape)
```

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    
    return g
```

```python
def compute_cost(X, y, w, b, *argv):
    m, n = X.shape
    
    total_cost = 0
    
    for i in range(m):
        z = b
        for j in range(n):
            z += w[j] * X[i][j]
        f_wb = sigmoid(z)
        loss = (-y[i] * np.log(f_wb)) - ((1 - y[i]) * np.log(1 - f_wb))
        total_cost += loss
        
    total_cost = total_cost / m 

    return total_cost
```

```python
def compute_cost_reg(X, y, w, b, lambda_ = 1):
    m, n = X.shape
    
    cost_without_reg = compute_cost(X, y, w, b) 
    
    reg_cost = 0.
    
    for i in range(n):
        reg_cost += w[i] * w[i]
        
    reg_cost = lambda_ / (2 * m) * reg_cost
    
    total_cost = cost_without_reg + reg_cost

    return total_cost
```

```python
def compute_gradient(X, y, w, b, *argv): 
    m, n = X.shape
    dj_dw = np.zeros(w.shape)
    dj_db = 0.

    for i in range(m):
        z_wb = b
        for j in range(n): 
            z_wb += w[j] * X[i][j]
        f_wb = sigmoid(z_wb)
        
        dj_db += f_wb - y[i]
        
        for j in range(n):
            dj_dw[j] += (f_wb - y[i])* X[i][j]
            
    dj_dw = dj_dw / m
    dj_db = dj_db / m

        
    return dj_db, dj_dw
```

```python
def compute_gradient_reg(X, y, w, b, lambda_ = 1): 
    m, n = X.shape
    
    dj_db, dj_dw = compute_gradient(X, y, w, b)

    for i in range(n):
        dj_dw[i] += lambda_ / m * w[i]        
        
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters, lambda_):
    m = len(X)
    J_history = []
    w_history = []
    
    for i in range(num_iters):
        dj_db, dj_dw = gradient_function(X, y, w_in, b_in, lambda_)
        w_in = w_in - alpha * dj_dw               
        b_in = b_in - alpha * dj_db 
        if i<100000:
            cost =  cost_function(X, y, w_in, b_in, lambda_)
            J_history.append(cost)
        if i% math.ceil(num_iters/10) == 0 or i == (num_iters-1):
            w_history.append(w_in)
            print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")

    return w_in, b_in, J_history, w_history
```

```python
def plot_decision_boundary(w, b, mapped_X, y_train):
    """
    Plot the decision boundary of a linear classifier.

    Parameters:
    w (numpy array): Weights of the linear classifier
    b (float): Bias of the linear classifier
    mapped_X (numpy array): Mapped features of the training data
    y_train (numpy array): Labels of the training data
    """
    # Create a meshgrid of points to evaluate the decision boundary
    x_min, x_max = mapped_X[:, 0].min() - 1, mapped_X[:, 0].max() + 1
    y_min, y_max = mapped_X[:, 1].min() - 1, mapped_X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

    # Evaluate the decision boundary
    Z = np.dot(np.c_[xx.ravel(), yy.ravel()], w.reshape(-1, 1)) + b
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.contour(xx, yy, Z, [0], colors='k')

    # Plot the training data
    plt.scatter(mapped_X[:, 0], mapped_X[:, 1], c=y_train, cmap=plt.cm.bwr)
```

```python
def predict(X, w, b):
    m, n = X.shape

    p = np.zeros(m)

    for i in range(m):
        z_wb = b
        for j in range(n):
            z_wb += w[j] * X[i][j]

        f_wb = sigmoid(z_wb)

        if f_wb >= 0.5:
            p[i] = 1
        else:
            p[i] = 0

    return p
```

```python
#Compute accuracy on the training set
p = predict(mapped_X, w, b)

print('Train Accuracy: %f'%(np.mean(p == y_train) * 100))
```
